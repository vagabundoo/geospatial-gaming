# Visualization

These 50 companies visualized below meet 3 of the criteria:

- Within 2km of a starbucks
- They are either gaming or design companies
- They raised over 1M USD in funding

```python
import geopandas as gpd
import pandas as pd
from cartoframes.viz import Map, Layer
from cartoframes.viz.helpers import size_continuous_layer
from cartoframes.viz.widgets import histogram_widget
df = pd.read_csv('../output/companies-gaming-highlyfunded-starbucks.csv')
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
Map(Layer(gdf,'color:green'))
```
